#### **객체(Object)** 
객체는 (1) 속성(Attribute)과 (2) 동작(Method)이 존재하는, 실존 - 메모리 공간이 할당되어 있는 - 하는 것을 의미합니다. 이렇게 보면 잘 이해가 되지 않으니 아래의 예시를 봅시다. 
> **예**
> * 사용자가 자판기에서 커피를 뽑아마신다. 
> * (1) **사용자**가 **자판기**에 **돈을 넣는다**.
> (2) **자판기**가 **동전을 처리**하고,**커피를 내논다**. 
> (3) **사용자**가 **자판기**에서 **커피를 꺼낸다**. 

위의 예시에서, **객체**는 사용자와 자판기를 의미합니다. 그리고 돈을 넣는다, 커피를 꺼낸다/동전을 처리한다, 커피를 내논다는 사용자와 자판기의 **동작(Method)**을 말하고, 이외에 사용자의 지갑 사정, 사용자의 이름, 자판기의 고유 번호 등이 각 객체가 가지는 **속성(Attribute)**이 됩니다. 

프로그래밍에서 흔히 사용되는 객체의 종류로는 이후에 배울 **정수형(Integer)**과 **문자형(String)**등이 있고, 이외에도 사용자의 필요에 따라 원하는 속성/동작을 포함한 객체를 정의하기도 합니다. 앞에서도 말했듯이 객체의 경우에는 속성과 동작을 공유하는데요, 이게 무슨 말인지 아래의 코드를 통해 봅시다
  

````python
# Integer Object
x = 1
y = 2
````

````python
print(x + y)
print(x - y)
print(x * y)
print(x / y)
````

### 자료형
* **자료형이란?** 
 자료형이란 변수의 종류를 말합니다. 정수형, 실수형, 문자형등 다양한 유형이 있으며, 지금 당장은 이 정도만 구분할 줄 알아도 충분합니다.

* **기본 자료형(Primitive Data Type)** 정수형, 실수형, 문자형, 논리형 등이 있으며, 더 복잡한 자료형을 구성하는 기본 단위가 됩니다.

### +) 시간이 나면 해봅시다 : String Formatting
* **f-string** :`f"stringblahblah{var1} and {var2}"`와 같은 식으로 문자열에 쉽게 값을 대입할 수 있습니다. 이외에도 f-string의 경우 다양한 형태로 문자열의 포맷을 설정해줄 수 있다는 것이 장점입니다. 
>**포멧 설정 방법**
>* `{[varname]:[정렬 방향, <(왼쪽 정렬) 또는 >(오른쪽 정렬).[소숫점 precision][datatype - s(string), d(digit), f(floating point)]}`의 형태로 나타내며, 이중 일부를 생략해도 무방
>* 정렬의 경우 숫자는 오른쪽, 문자는 왼쪽이 기본 설정!




````python
name = "박수빈"
birthday = "2/14"
score = 96.75
print(f"Hello, my name is {name}. My birthday is {birthday}.")
print(f"My score is {score:.0f}.")

# TODO: 콘솔 입력으로 이름, 학번 받아서 
# 안녕하세요, 저는 {이름} 이고, 학번은 {}입니다. 
````

````python
name = "subeen"
height = 160
age = 22
gpa = 3.3333

# name이 12칸을 차지하도록, value가 7칸을 차지하도록!
print(f"{name:12s}: {height:7d}cm") 

# name이 12칸을 차지하면서 우측 정렬, value가 7칸을 차지면서 왼쪽 정렬 되도록!
print(f"{name:>12s}: {height:<7d}")

# gpa가 8자리를 차지하면서 소수 둘째자리까지 나타나도록!
print(f"gpa:{gpa:8.2f}")
````

## 2.3 불(Boolean) 자료형
**불(Bool, Boolean)** 자료형은 참과 거짓을 나타내는 자료형입니다. 값이 두 가지 밖에 없는 자료형이기 때문에 크게 주의할 것은 없지만, 언제나 **대문자로 시작**해야만 불 자료형으로 인식된다는 것을 명심하세요!(`True`, `False`는 되지만 `true`, `false`는 안됩니다ㅠㅠㅠㅠ) 그럼 이제 아래의 코드를 통해서 불 자료형을 반환하는 **논리 연산자(Logic Operator)**를 한번 봅시다! 아래 코드들의 결과값을 예측해보세요!



````python
# ==
#print("hello" == "Hello")

# != : 같지 않다
# print (3 != 3)
# print ("hello" != "hello")

# <, >, >=, <=
# print(3<=4)
# print(35/7 >= 5)

# and, or, not
print(not ("Hello" == "Hello"))
print((3 > 5) and (4 != 3))

# 3이 4보다 크거나 5가 7보다 작거나 같다
# a가 5보다 크면서 b가 7보다 크다 (콘솔 입출력)
````

Expression 뿐만 아니라 **자료형 자체**도 불 값을 가집니다. 자료형의 bool 값을 확인해주기 위해서는 integer, string 등의 자료형을 bool()함수 안에 넣어 bool 자료형으로 변환해주면 됩니다. 

````python
# 자료형 자체의 boolean 값
print(bool(""))
print(bool(0))
print(bool([]))
print(bool("False"))
print(bool(1))
print(bool(-1))
````

이쯤에서 잠시 **형변환(Type Casting)**이 무엇인지 볼까요? **형변환**이란 한 자료형을 다른 자료형으로 바꾸어주는 행위를 말합니다. 사용자 정의형 자료형 이외에 이번 수업에서 배운 기본 자료형, 복합 자료형의 경우 사이의 형변환이 어느 정도 자유롭습니다. 그렇지만 합리적으로 생각해보았을 때 어떤 결과값이 나올지 잘 예측되지 않는다거나(예: `int("hello)`), 데이터를 잃을 우려가 있다(예: `int(3.41)`)고 생각되는 경우에는 형변환을 자제하는 것이 좋습니다. 

````python
# 형변환(Type Casting)
a = 3
b = 5
print(a + b)
print(str(b)*3) #str(variable) : 문자열
c = "03"
print(c)
print(int(c)) #int(variable) : 정수형 
print(bool(c)) #bool(variable) : 불리안
print(float(3)) #float : 실수형
d = 3.6
print(int(d))
````

## 2.6 딕셔너리(Dictionary) 자료형 
---
* **딕셔너리(Dictionary) 자료형**은 **key**와 **value**를 가지는 자료형입니다. key와 value는 서로 대응 관계에 있는 값인데요, 예를 들어 설명하자면 학교-서울대, 학과-지구환경과학부와 같은 대응 관계에서 학교와 학과가 key, 서울대, 지구환경과학부가 값이 됩니다. 
* 지금까지 배운 리스트와 튜플의 경우 **인덱스**를 이용해 값에 접근했지만, 딕셔너리 자료형에서는 이와 달리 값에 순차적으로 접근하는 대신 **key**를 통해 데이터에 접근합니다. 이렇게만 들으면 잘 와닿지 않으니 아래의 코드를 함께 보며 이해해봅시다. 
* 주의사항
  * 딕셔너리 자료형의 key 값으로는 **immutable type**만 가능합니다. immutable type의 경우 데이터의 원형을 변형시킬 수 없는 경우를 말하는데, tuple이 바로 그 대표적인 예시입니다. 다만 깊게 들어가면 복잡해지니 지금으로서는 integer, string, float 정도만 key 값으로 사용할 수 있다고 생각해둡시다. 
  * 딕셔너리 key는 중복 불가! 나중에 입력된 값만 저장

````python
studentInfo = {'name':'박수빈', 'major':['컴공', '경영', '지환'], 'age':22}
print(studentInfo['name'])
print(studentInfo['major'])
print(studentInfo['age'])


studentInfo['major'].remove('지환')
print(studentInfo['major'])
````

### 딕셔너리 다루기: 빈 딕셔너리 만들기, 딕셔너리 초기값 설정하기, 쌍 추가하기, 삭제하기
* `dict()`: 빈 딕셔너리를 형성
* `dict([[key1, val1], [key2, val2], [key3, val3]])`: 초기 값이 지정된 딕셔너리 생성
* `dic1[key1] = [val1]` : 딕셔너리에 key-value pair 삽입
* `del dic1[key1]`: key-value pair 삭제



````python
# 빈 딕셔너리 선언
example = dict() 
example2 = {}
# print(example)
# print(example2)

# 초기값을 가진 딕셔너리 만들기
example3 = {"key1": "value1", "key2": "value2"}
example4 = dict([("key1", "value1"), ("key2", "value2")])
# print(example3)
# print(example4)

# 쌍 추가하기: dicname[key] = value
example[1] = "practice"
example[2] = "makes"
example[3] = "perfect"
print(example)

# key에 대응되는 value 값 수정하기
example[1] = "manner"
example[2] = "maketh"
example[3] = "man"
print(example)

# 요소 삭제하기 : del dictname[key]
del example[2]
print(example)
````

### 딕셔너리 관련 함수
* `dict.keys()`: dict의 key만을 모아서 dict_keys 형 데이터를 반환
* `dict.items()`: dict의 (key, item) 튜플을 모아 dict_items 형 데이터를 반환
* `dict.clear()`: dictionary를 빈 dictionary로 돌려놓음
* `dict.get(key)`: key에 해당하는 value를 돌려줌.(없을 시 none 반환! <-> dict[key]의 경우 없는 key를 찾게 되면 오류를 일으키지만 이 함수의 경우 그렇지 않음)

````python
print(example.keys())
print(example.items())
# example.clear()
print(example.get(1))
# print(example[2])
print(example.get(2))
````
